```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('preprocessed_train.csv')
```

```python
df['Date'] = pd.to_datetime(df.Date)
df['Year'] = df['Date'].apply(lambda x: x.year)
# get one store
store_df = df[(df.StoreID == 1600)]
# store_df = store_df[store_df.DayOfWeek == 5]
# store_df = store_df[store_df.HasPromotions == 0]
```

```python
dates = pd.to_datetime(store_df.Date)
sales = store_df.NumberOfSales.values
```

```python
fig = plt.figure(figsize=(24, 24))
ax = plt.axes()
ax.plot(dates, sales)
```
