```python
import pandas as pd
import numpy as np
import datetime as dt
from IPython.display import clear_output
```

```python
df = pd.read_csv('preprocessed_train.csv')
df.head()
```

```python
def split(df):
    start = pd.to_datetime('1/03/2017', format='%d/%m/%Y')
    stop = pd.to_datetime('1/05/2017', format='%d/%m/%Y')
    train = pd.concat([df[df.Date < start],df[df.Date >= stop]])
    test = df.drop(train.index)
    return train, test
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
df['Year']=df['Date'].dt.year
```

```python
df, test = split(df)
```

```python
for store in df.StoreID.unique().tolist():
# for store in range(1000,1002):
# store = 1000
    print('store ', store)
    mask = (df.StoreID == store)
    store_df = df[mask]

#     avg daily sales
    available_months = store_df.Month.unique().tolist()
    available_years = store_df.Year.unique().tolist()
#     print(available_months,available_years)

    total_sales = sum(store_df.NumberOfSales.tolist())
    total_open_days = store_df.Date.count()
    df.loc[mask, 'daily_sales'] = total_sales/total_open_days
#     print('daily_sales', total_sales/total_open_days)
    
#     avg sales for each month (BASED ON specific MONTH, not just average of all months)
    for m in available_months:
        df.loc[((mask) & (df.Month == m)), 'month_avg_sales'] = sum(store_df[(store_df.Month == m)].NumberOfSales.tolist())/len(store_df[(store_df.Month == m)].Year.unique().tolist())
    
# avg yearly sales
    df.loc[mask , 'yearly_sales'] = sum(store_df.NumberOfSales.tolist())/len(store_df['Year'].unique().tolist())
```

```python
# Get columns with NaN values
null_cols = []
print('Features with NaN:')
for col in df.columns:
    if df[col].isnull().values.any():
        print(col)
        null_cols.append(col)
```
